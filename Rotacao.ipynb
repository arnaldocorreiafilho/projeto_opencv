```
from PIL import Image
import cv2 as cv
from matplotlib import pyplot as plt  
import numpy as np
% matplotlib inline
```

Importação das bibliotecas que serão utilizadas para o processamento das imagens.

```
img = cv.imread("image1.png",0)
plt.imshow(img,cmap=plt.cm.gray)
```

Leitura e exibição da imagens em tons de cinza.

```
linhas , colunas = img.shape
```

Pegar o quantidade de linhas e colunas através do método shape.

```
mat = cv.getRotationMatrix2D((linhas/2,colunas/2),15,1)

```

Pega a matriz que fará a operação na imagem onde se passa o centro da imagen o ãngulo e o tamanho que ela ficará no caso foi colocado o número 1 para dizer que a imagem ficará no tamanho original.

```
imgrot = cv.warpAffine(img,mat,(linhas,colunas))
plt.imshow(imgrot,cmap=plt.cm.gray)
```

Com a função warpAffine gerará uma nova imagem já feita a operação de rotação. Seus parâmetros é a imagen original e operação de rotação e o tamanho da imagem através das linhas e colunas obtidas através da função shape.
